```python
def onehot_encode(df, column):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix = column)
    df = pd.concat([df,dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
df2 = df1.copy()
```

```python
df2['Bathroom'] = df1['Bathroom'].fillna(df1['Bathroom'].mode()[0])
df2
```

```python
df2['Furnishing'] = df1['Furnishing']
```

```python
df2['Parking'] = df1['Parking'].fillna(df1['Parking'].mode()[0])
```

```python
df2.isna().sum()
```

```python
df2['Type'] = df1['Type'].fillna(df1['Type'].mode()[0])
```

```python
df2.isna().sum()
```

```python
#Handling categorical data

pd.get_dummies(df2.Furnishing)
```

```python
#binary encoding
df2['Status'] = df2['Status'].replace({
    'Ready_to_move':0, 
    'Almost_ready':1
})

df2['Transaction'] = df2['Transaction'].replace({
    'New_Property':0, 
    'Resale':1
})

df2['Type'] = df2['Type'].replace({
    'Builder_Floor':0, 
    'Apartment':1
})
df2
```

```python
df3 = onehot_encode(df2, column='Furnishing')
df3
```

```python
df3.isna().sum()
```

```python
df4 = df3.drop(['Per_Sqft','Locality'], axis=1)
df4
```

```python
df4[df4.Area/df4.BHK<50]
```

```python
df5 = df4[~(df4.Area/df4.BHK<100)]
df5
```

```python
df5 = df5[(df5['Area']<=5000)]
df5
```

```python
df5 = df5[df5['BHK']<7]
```

```python
df5 = df5[~(df5['Bathroom']==6)]
df5.shape
```

```python
df6 = df5.drop_duplicates()
df6.shape
```

```python
{column: list(df6[column].unique()) for column in df6.select_dtypes('object').columns}
```

```python
y = df6['Price']
X = df6.drop('Price', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
#Splitting the data to train the model


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
models = {
    "                     Linear Regression": LinearRegression(),
    " Linear Regression (L2 Regularization)": Ridge(),
    " Linear Regression (L1 Regularization)": Lasso(),
    "                   K-Nearest Neighbors": KNeighborsRegressor(),
    "                         Decision Tree": DecisionTreeRegressor(),
    "                         Random Forest": RandomForestRegressor(),
}

for name,model in models.items():
    model.fit(X_train, y_train)
    print(name+ " trained.")
```

```python
for name,model in models.items():
    print(name+ "  Score: {:.5f}".format(model.score(X_test,y_test)))
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
import pickle
with open('DHP_model.pickle','wb') as f:
    pickle.dump(rf,f)
```
